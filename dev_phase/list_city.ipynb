```python
import pandas as pd
import json
```

```python
# Loại bỏ tham số lines=True
df = pd.read_json('city.list.json')

# Kiểm tra dữ liệu
print(df.head())
```

```python
df_vn = df[df['country'] == 'VN'].copy()
```

```python
df_vn['coord']
```

```python
df_coord_new = df_vn['coord'].apply(pd.Series)

# Gán lại tên cột cho dễ đọc
df_coord_new = df_coord_new.rename(columns={'lon': 'longitude', 'lat': 'latitude'})

# Kết hợp DataFrame gốc và DataFrame mới
df_vn = pd.concat([df_vn.drop('coord', axis=1), df_coord_new], axis=1)

print(df_vn)
```

```python
# Set options to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
valid_names = [
    "Huyện Ðiện Biên",
    "Tỉnh Cao Bằng",
    "Thành phố Hồ Chí Minh",
    "Tỉnh Nghệ An",
    "Tỉnh Ninh Bình",
    "Tỉnh Ninh Thuận",
    "Tỉnh Sóc Trăng",
    "Tỉnh Trà Vinh",
    "Tỉnh Tuyên Quang",
    "Tỉnh Vĩnh Long",
    "Tỉnh Yên Bái",
    "Tỉnh Lào Cai",
    "Tỉnh Tiền Giang",
    "Tỉnh Thừa Thiên-Huế",
    "Tỉnh Kon Tum",
    "Tỉnh Thanh Hóa",
    "Tỉnh Thái Bình",
    "Tỉnh Tây Ninh",
    "Tỉnh Sơn La",
    "Tỉnh Quảng Trị",
    "Tỉnh Quảng Ninh",
    "Tỉnh Quảng Ngãi",
    "Tỉnh Quảng Bình",
    "Tỉnh Phú Yên",
    "Tỉnh Long An",
    "Tỉnh Hòa Bình",
    "Tỉnh Lạng Sơn",
    "Tỉnh Lâm Ðồng",
    "Tỉnh Kiến Giang",
    "Tỉnh Khánh Hòa",
    "Tỉnh Hà Tĩnh",
    "Tỉnh Hà Giang",
    "Tỉnh Gia Lai",
    "Thủ Ðô Hà Nội",
    "Tỉnh Cần Thơ",
    "Thành Phố Hải Phòng",
    "Tỉnh Bình Thuận",
    "Tỉnh Ðồng Tháp",
    "Tỉnh Ðồng Nai",
    "Tỉnh Ðắc Lắk",
    "Tỉnh Bà Rịa-Vũng Tàu",
    "Tỉnh Bình Ðịnh",
    "Tỉnh Bến Tre",
    "Tỉnh An Giang",
    "Tỉnh Bắc Ninh",
    "Tỉnh Bắc Giang",
    "Tỉnh Ðà Nẵng",
    "Tỉnh Bình Dương",
    "Tỉnh Bình Phước",
    "Tỉnh Thái Nguyên",
    "Tỉnh Quảng Nam",
    "Tỉnh Phú Thọ",
    "Tỉnh Nam Ðịnh",
    "Tỉnh Hà Nam",
    "Tỉnh Bắc Kạn",
    "Tỉnh Bạc Liêu",
    "Tỉnh Cà Mau",
    "Tỉnh Hải Dương",
    "Tỉnh Hưng Yên",
    "Tỉnh Vĩnh Phúc",
]
df_vn_new = df_vn[df_vn['name'].isin(valid_names)].copy()
```

```python
df_vn_new
```

```python
df_vn_new["name"] = (
    df_vn_new["name"]
    .str.replace(r"^Tỉnh\s+", "", regex=True)
    .str.replace(r"^Thành[ ]?Phố\s+", "", regex=True)
    .str.replace(r"^Thủ[ ]?Ðô\s+", "", regex=True)
    .str.replace(r"^Thành[ ]?phố\s+", "", regex=True)
    .str.replace(r"^Huyện\s+", "", regex=True)
    .str.strip()
)
```

```python
df_vn_new
```

```python
# mapping các tên sai -> tên chuẩn
name_mapping = {
    "Lâm Ðồng": "Lâm Đồng",
    "Bình Ðịnh": "Bình Định",
    "Ðồng Tháp": "Đồng Tháp",
    "Ðồng Nai": "Đồng Nai",
    "Ðắc Lắk": "Đắk Lắk",
    "Nam Ðịnh": "Nam Định",
    "Ðiện Biên": "Điện Biên",
    "Ðà Nẵng": "Đà Nẵng",
    "Kiến Giang": "Kiên Giang",
    "Hồ Chí Minh": "TP. Hồ Chí Minh",
    "Bà Rịa-Vũng Tàu": "Bà Rịa - Vũng Tàu",
}

# áp dụng thay thế
df_vn_new["name"] = df_vn_new["name"].replace(name_mapping)

# kiểm tra lại kết quả
print(df_vn_new["name"].unique())
```

```python
df_vn_new.drop(columns='country', inplace=True)
```

```python
df_vn_new.to_csv('city.csv', index=False, encoding='utf-8-sig')
```
